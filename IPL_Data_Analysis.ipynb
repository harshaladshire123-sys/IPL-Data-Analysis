```python
# IPL Cricket Data Analysis Project

# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load Dataset
matches = pd.read_csv("matches.csv")
deliveries = pd.read_csv("deliveries.csv")

# Show First 5 Rows
print(matches.head())

# -----------------------------
# Top 10 Run Scorers
# -----------------------------

top_batsmen = deliveries.groupby('batsman')['batsman_runs'].sum()

top_batsmen = top_batsmen.sort_values(ascending=False).head(10)

print(top_batsmen)

# Plot Graph
plt.figure(figsize=(10,5))
top_batsmen.plot(kind='bar')

plt.title("Top 10 IPL Run Scorers")
plt.xlabel("Players")
plt.ylabel("Runs")

plt.xticks(rotation=45)

plt.show()

# -----------------------------
# Most Winning Teams
# -----------------------------

team_wins = matches['winner'].value_counts().head(10)

print(team_wins)

# Plot Pie Chart
plt.figure(figsize=(8,8))

team_wins.plot(kind='pie', autopct='%1.1f%%')

plt.title("Most Successful IPL Teams")

plt.ylabel("")

plt.show()

# -----------------------------
# Toss Decision Analysis
# -----------------------------

toss = matches['toss_decision'].value_counts()

print(toss)

# Plot Bar Chart
plt.figure(figsize=(6,4))

toss.plot(kind='bar')

plt.title("Toss Decision Analysis")
plt.xlabel("Decision")
plt.ylabel("Count")

plt.show()

# -----------------------------
# Highest Team Scores
# -----------------------------

high_scores = deliveries.groupby(
    ['match_id', 'batting_team']
)['total_runs'].sum()

high_scores = high_scores.sort_values(
    ascending=False
).head(10)

print(high_scores)

# -----------------------------
# Save Cleaned Data
# -----------------------------

matches.to_csv("cleaned_matches.csv", index=False)

print("Project Completed Successfully")
```
